```python
df_x=df.drop(columns=['Attribute_61'])
df_y=df[['Attribute_61']]
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
le.fit(df_y)
y=le.transform(df_y)
y
```

```python
x=np.array(df_x)
```

```python
x
```

```python
from sklearn.metrics import f1_score
from sklearn.model_selection import train_test_split
def maxf1_score(clf,x,y):
    maxf=0
    rs=0
    for r_state in range(42,100):
        x_train,x_test,y_train,y_test=train_test_split(x, y,random_state = r_state,test_size=0.20,stratify=y)
        clf.fit(x_train,y_train)
        y_pred=clf.predict(x_test)
        tmp=f1_score(y_test,y_pred)
        if tmp>maxf:
            maxf=tmp
            rs=r_state
    print("maximum f1_score is at random state :",rs," and it is :",maxf)
```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
lg=LogisticRegression()
maxf1_score(lg,x,y)
```

```python
import numpy as np
from sklearn.model_selection import StratifiedKFold
f1_score_logistic=np.array([])
skf = StratifiedKFold(n_splits=13)
for train_index, test_index in skf.split(x, y):
    x_train, x_test = x[train_index], x[test_index]
    y_train, y_test = y[train_index], y[test_index]
    lg.fit(x_train,y_train)
    y_pred=lg.predict(x_test)
    tmp=f1_score(y_test,y_pred)
    f1_score_logistic=np.append(f1_score_logistic,[tmp])
print("mean f1 score for logsistic after stratified cross validation : ",np.mean(f1_score_logistic))
print("standard deviation for logistic from mean f1 score is : ",np.std(f1_score_logistic))
print("array of f1 scores : ",f1_score_logistic)
```

```python
#decision tree classifier
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier()
maxf1_score(dtc,x,y)
```

```python
f1_score_dtc=np.array([])
skf = StratifiedKFold(n_splits=13)
for train_index, test_index in skf.split(df_x, y):
    x_train, x_test = x[train_index], x[test_index]
    y_train, y_test = y[train_index], y[test_index]
    dtc.fit(x_train,y_train)
    y_pred=dtc.predict(x_test)
    tmp=f1_score(y_test,y_pred)
    f1_score_dtc=np.append(f1_score_dtc,[tmp])
print("mean f1 score for decision after stratified cross validation : ",np.mean(f1_score_dtc))
print("standard deviation for decision from mean f1 score is : ",np.std(f1_score_dtc))
print("array of f1 scores : ",f1_score_dtc)
```

```python
#KN CLassifier
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
knc=KNeighborsClassifier()
neighbors={"n_neighbors":range(1,30)}
clf = GridSearchCV(knc, neighbors, cv=6,scoring="f1")
clf.fit(x,y)
clf.best_params_
```

```python
knc=KNeighborsClassifier(n_neighbors=13)
maxf1_score(knc,x,y)
```

```python
#SVC
from sklearn.svm import SVC
svc=SVC()
parameters={"kernel":["linear", "poly", "rbf"],"C":[0.001,0.01,0.1,1,10]}
clf = GridSearchCV(svc, parameters, cv=5,scoring="f1")
clf.fit(df_x,y)
clf.best_params_
```
